### Degree Distribution Analysis

**Why this matters**: Degree distribution reveals the structure of social networks. Real social networks typically follow a power-law distribution (scale-free property), meaning most nodes have few connections while a few "hub" nodes have many connections. This affects which algorithms will work well.

```python
# Calculate degree statistics
degrees = dict(G.degree())
degree_values = list(degrees.values())

print("DEGREE STATISTICS")
print("="*70)
print(f"Average degree: {np.mean(degree_values):.2f}")
print(f"Median degree: {np.median(degree_values):.2f}")
print(f"Standard deviation: {np.std(degree_values):.2f}")
print(f"Min degree: {min(degree_values)}")
print(f"Max degree: {max(degree_values)}")
print(f"\nDegree range: {max(degree_values) - min(degree_values)}")
print(f"Coefficient of variation: {np.std(degree_values)/np.mean(degree_values):.2f}")
print("="*70)

# Top nodes by degree
print("\nTop 10 nodes by degree:")
sorted_degrees = sorted(degrees.items(), key=lambda x: x[1], reverse=True)
for node, deg in sorted_degrees[:10]:
    print(f"  Node {node}: {deg} connections")
```

```python
# Visualize degree distribution
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Histogram
axes[0].hist(degree_values, bins=20, color='steelblue', edgecolor='black', alpha=0.7)
axes[0].axvline(np.mean(degree_values), color='red', linestyle='--', linewidth=2, label=f'Mean: {np.mean(degree_values):.2f}')
axes[0].axvline(np.median(degree_values), color='orange', linestyle='--', linewidth=2, label=f'Median: {np.median(degree_values):.2f}')
axes[0].set_xlabel('Degree', fontweight='bold')
axes[0].set_ylabel('Frequency', fontweight='bold')
axes[0].set_title('Degree Distribution (Histogram)', fontweight='bold', fontsize=12)
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Log-log plot to check for power-law
degree_counts = Counter(degree_values)
degrees_x = sorted(degree_counts.keys())
counts_y = [degree_counts[d] for d in degrees_x]

axes[1].loglog(degrees_x, counts_y, 'o-', markersize=8, linewidth=2, color='darkgreen')
axes[1].set_xlabel('Degree (log scale)', fontweight='bold')
axes[1].set_ylabel('Frequency (log scale)', fontweight='bold')
axes[1].set_title('Degree Distribution (Log-Log)', fontweight='bold', fontsize=12)
axes[1].grid(True, which="both", alpha=0.3)

# Cumulative distribution
sorted_degrees_vals = sorted(degree_values, reverse=True)
axes[2].plot(range(len(sorted_degrees_vals)), sorted_degrees_vals, linewidth=2.5, color='purple')
axes[2].set_xlabel('Node Rank', fontweight='bold')
axes[2].set_ylabel('Degree', fontweight='bold')
axes[2].set_title('Degree Rank Plot', fontweight='bold', fontsize=12)
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n📊 INTERPRETATION:")
print("   - The log-log plot helps identify if the network follows a power-law (straight line = power-law)")
print("   - High variance in degrees suggests hub nodes exist (important for community structure)")
print("   - The rank plot shows how concentrated connections are among top nodes")
```

### Network Centrality Analysis

**Why multiple centrality measures**: Different centrality metrics capture different notions of "importance":
- **Degree**: Local popularity (direct connections)
- **Betweenness**: Bridge nodes connecting different communities
- **Closeness**: Global reach (can access others quickly)
- **Eigenvector**: Connected to other important nodes (quality over quantity)

Comparing these reveals different structural roles nodes play.

```python
# Calculate centrality measures
print("Computing centrality measures...")
degree_cent = nx.degree_centrality(G)
betweenness_cent = nx.betweenness_centrality(G)
closeness_cent = nx.closeness_centrality(G)
eigenvector_cent = nx.eigenvector_centrality(G, max_iter=1000)

# Create DataFrame for analysis
centrality_df = pd.DataFrame({
    'Node': list(G.nodes()),
    'Degree_Centrality': [degree_cent[n] for n in G.nodes()],
    'Betweenness_Centrality': [betweenness_cent[n] for n in G.nodes()],
    'Closeness_Centrality': [closeness_cent[n] for n in G.nodes()],
    'Eigenvector_Centrality': [eigenvector_cent[n] for n in G.nodes()],
    'Actual_Degree': [degrees[n] for n in G.nodes()],
    'Community': [ground_truth[n] for n in G.nodes()]
})

print("\nCENTRALITY STATISTICS")
print("="*70)
print(centrality_df.describe()[['Degree_Centrality', 'Betweenness_Centrality', 
                                  'Closeness_Centrality', 'Eigenvector_Centrality']])

print("\n\nTop 5 nodes by each centrality measure:")
print("="*70)
for col in ['Degree_Centrality', 'Betweenness_Centrality', 'Closeness_Centrality', 'Eigenvector_Centrality']:
    print(f"\n{col}:")
    top5 = centrality_df.nlargest(5, col)[['Node', col, 'Community']]
    print(top5.to_string(index=False))
```

```python
# Correlation between centrality measures
# Why: Understanding correlation helps identify if measures capture similar/different information

cent_cols = ['Degree_Centrality', 'Betweenness_Centrality', 'Closeness_Centrality', 'Eigenvector_Centrality']
correlation_matrix = centrality_df[cent_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.3f', cmap='coolwarm', 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Between Centrality Measures', fontweight='bold', fontsize=14, pad=20)
plt.tight_layout()
plt.show()

print("\n📊 INTERPRETATION:")
print("   - High correlation (>0.8): Measures capture similar information")
print("   - Low correlation (<0.5): Measures reveal different aspects of node importance")
print("   - This informs which features to use in beyond-course GNN models")
```

### Clustering and Community Structure

**Why clustering coefficient matters**: High clustering indicates tightly-knit groups (friends-of-friends are friends), which is characteristic of social networks and suggests strong community structure.

**Transitivity**: Global measure of clustering across the entire network.

```python
# Calculate clustering coefficients
clustering_coef = nx.clustering(G)
avg_clustering = nx.average_clustering(G)
transitivity = nx.transitivity(G)

# Triangle count
triangles = nx.triangles(G)
total_triangles = sum(triangles.values()) // 3

print("CLUSTERING ANALYSIS")
print("="*70)
print(f"Average clustering coefficient: {avg_clustering:.4f}")
print(f"Network transitivity: {transitivity:.4f}")
print(f"Total triangles: {total_triangles}")
print(f"\nClustering coefficient statistics:")
print(f"  Mean: {np.mean(list(clustering_coef.values())):.4f}")
print(f"  Std: {np.std(list(clustering_coef.values())):.4f}")
print(f"  Min: {min(clustering_coef.values()):.4f}")
print(f"  Max: {max(clustering_coef.values()):.4f}")

# Compare with random graph
G_random = nx.erdos_renyi_graph(G.number_of_nodes(), nx.density(G), seed=42)
avg_clustering_random = nx.average_clustering(G_random)
print(f"\nComparison with Erdős-Rényi random graph:")
print(f"  Social network clustering: {avg_clustering:.4f}")
print(f"  Random graph clustering: {avg_clustering_random:.4f}")
print(f"  Ratio (social/random): {avg_clustering/avg_clustering_random:.2f}x")
print("\n📊 High ratio suggests strong community structure!")
print("="*70)
```

### Community Detection (Course Technique: Louvain Algorithm)

**Algorithm choice rationale**: 
- **Louvain** is an unsupervised community detection algorithm that maximizes modularity
- It's fast and scales well (O(n log n))
- No need to specify number of communities in advance
- Industry standard for large networks

**Modularity** measures how well the network is divided into communities (range: -0.5 to 1, higher is better)

```python
# Apply Louvain community detection
detected_communities = community_louvain.best_partition(G)

# Calculate modularity
modularity = community_louvain.modularity(detected_communities, G)

# Community sizes
community_sizes = Counter(detected_communities.values())
num_communities = len(community_sizes)

print("COMMUNITY DETECTION RESULTS (LOUVAIN)")
print("="*70)
print(f"Number of communities detected: {num_communities}")
print(f"Modularity score: {modularity:.4f}")
print(f"\nCommunity size distribution:")
for comm_id, size in sorted(community_sizes.items(), key=lambda x: x[1], reverse=True):
    print(f"  Community {comm_id}: {size} nodes ({size/G.number_of_nodes()*100:.1f}%)")

# Compare with ground truth
# Map ground truth to numeric labels
gt_mapping = {label: idx for idx, label in enumerate(set(ground_truth.values()))}
ground_truth_numeric = {node: gt_mapping[label] for node, label in ground_truth.items()}

# Calculate agreement using Normalized Mutual Information
from sklearn.metrics import normalized_mutual_info_score, adjusted_rand_score

gt_labels = [ground_truth_numeric[n] for n in sorted(G.nodes())]
detected_labels = [detected_communities[n] for n in sorted(G.nodes())]

nmi = normalized_mutual_info_score(gt_labels, detected_labels)
ari = adjusted_rand_score(gt_labels, detected_labels)

print(f"\nComparison with ground truth:")
print(f"  Normalized Mutual Information (NMI): {nmi:.4f}")
print(f"  Adjusted Rand Index (ARI): {ari:.4f}")
print(f"  Number of ground truth communities: {len(set(ground_truth.values()))}")
print("\n📊 NMI/ARI close to 1 = strong agreement with ground truth")
print("="*70)
```

```python
# Visualize detected communities
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Layout for consistent visualization
pos = nx.spring_layout(G, seed=42, k=0.5, iterations=50)

# Plot 1: Detected communities
node_colors_detected = [detected_communities[n] for n in G.nodes()]
nx.draw(G, pos, ax=axes[0], node_color=node_colors_detected, 
        with_labels=True, node_size=500, cmap='Set3',
        font_size=9, font_weight='bold', edge_color='gray', width=1.5, alpha=0.9)
axes[0].set_title(f'Louvain Communities (Detected)\nModularity: {modularity:.3f} | Communities: {num_communities}', 
                  fontweight='bold', fontsize=12)

# Plot 2: Ground truth communities
node_colors_gt = [gt_mapping[ground_truth[n]] for n in G.nodes()]
nx.draw(G, pos, ax=axes[1], node_color=node_colors_gt,
        with_labels=True, node_size=500, cmap='Set3',
        font_size=9, font_weight='bold', edge_color='gray', width=1.5, alpha=0.9)
axes[1].set_title(f'Ground Truth Communities\nNMI: {nmi:.3f} | ARI: {ari:.3f}', 
                  fontweight='bold', fontsize=12)

plt.tight_layout()
plt.show()
```

### Path Analysis and Network Diameter

**Why this matters**: 
- **Average shortest path**: How many "hops" to connect any two people (small-world property)
- **Diameter**: Worst-case distance between any two nodes
- Social networks typically have short paths ("six degrees of separation")

```python
# Path analysis
avg_shortest_path = nx.average_shortest_path_length(G)
diameter = nx.diameter(G)
radius = nx.radius(G)

# Calculate all shortest paths (for distribution)
all_path_lengths = []
for source in G.nodes():
    path_lengths = nx.single_source_shortest_path_length(G, source)
    all_path_lengths.extend(path_lengths.values())

print("PATH AND DISTANCE ANALYSIS")
print("="*70)
print(f"Average shortest path length: {avg_shortest_path:.4f}")
print(f"Network diameter: {diameter}")
print(f"Network radius: {radius}")
print(f"\nShortest path distribution:")
print(f"  Min: {min([p for p in all_path_lengths if p > 0])}")
print(f"  Max: {max(all_path_lengths)}")
print(f"  Mean: {np.mean(all_path_lengths):.4f}")
print(f"  Median: {np.median(all_path_lengths):.4f}")

# Compare with random graph
avg_path_random = nx.average_shortest_path_length(G_random)
print(f"\nComparison with random graph:")
print(f"  Social network avg path: {avg_shortest_path:.4f}")
print(f"  Random graph avg path: {avg_path_random:.4f}")
print(f"\n📊 Similar path lengths suggest 'small-world' property")
print("="*70)
```

```python
# Visualize path length distribution
path_dist = Counter(all_path_lengths)

fig, ax = plt.subplots(figsize=(12, 6))
distances = sorted(path_dist.keys())
counts = [path_dist[d] for d in distances]

ax.bar(distances, counts, color='teal', edgecolor='black', alpha=0.7)
ax.axvline(avg_shortest_path, color='red', linestyle='--', linewidth=2.5, 
           label=f'Average: {avg_shortest_path:.2f}')
ax.set_xlabel('Shortest Path Length', fontweight='bold', fontsize=12)
ax.set_ylabel('Frequency', fontweight='bold', fontsize=12)
ax.set_title('Distribution of Shortest Path Lengths', fontweight='bold', fontsize=14)
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

### Assortativity and Homophily

**Degree assortativity**: Do high-degree nodes connect to other high-degree nodes?
- Positive assortativity (>0): "Rich connect with rich" (social networks)
- Negative assortativity (<0): "Hubs connect with peripheral nodes" (technological networks)

This affects information spread and network resilience.

```python
# Assortativity analysis
degree_assortativity = nx.degree_assortativity_coefficient(G)

print("ASSORTATIVITY ANALYSIS")
print("="*70)
print(f"Degree assortativity coefficient: {degree_assortativity:.4f}")

if degree_assortativity > 0.3:
    print("  → Strong positive assortativity: High-degree nodes connect to each other")
elif degree_assortativity > 0:
    print("  → Weak positive assortativity: Slight tendency for similar degrees to connect")
elif degree_assortativity > -0.3:
    print("  → Weak negative assortativity: High-degree nodes somewhat avoid each other")
else:
    print("  → Strong negative assortativity: High-degree nodes connect to low-degree nodes")

print("\n📊 Social networks typically show positive assortativity")
print("="*70)

# Visualize assortativity
# For each edge, plot (degree of node1, degree of node2)
edge_degrees = [(degrees[u], degrees[v]) for u, v in G.edges()]
degrees_u = [d[0] for d in edge_degrees]
degrees_v = [d[1] for d in edge_degrees]

fig, ax = plt.subplots(figsize=(10, 8))
ax.hexbin(degrees_u, degrees_v, gridsize=15, cmap='YlOrRd', mincnt=1)
ax.set_xlabel('Node Degree (u)', fontweight='bold', fontsize=12)
ax.set_ylabel('Node Degree (v)', fontweight='bold', fontsize=12)
ax.set_title(f'Degree Assortativity Visualization\nCoefficient: {degree_assortativity:.3f}', 
             fontweight='bold', fontsize=14)
plt.colorbar(ax.collections[0], ax=ax, label='Edge Count')
plt.tight_layout()
plt.show()
```

```python
# Data quality checks
print("DATA QUALITY ASSESSMENT")
print("="*70)

# 1. Check for self-loops
self_loops = list(nx.selfloop_edges(G))
print(f"Self-loops: {len(self_loops)}")
if self_loops:
    print(f"  ⚠ Warning: Found self-loops at nodes: {[u for u, v in self_loops]}")
else:
    print("  ✓ No self-loops (good!)")

# 2. Check for isolated nodes
isolated_nodes = list(nx.isolates(G))
print(f"\nIsolated nodes: {len(isolated_nodes)}")
if isolated_nodes:
    print(f"  ⚠ Warning: Found isolated nodes: {isolated_nodes}")
else:
    print("  ✓ No isolated nodes (good!)")

# 3. Check for multiple edges (only for MultiGraph)
if isinstance(G, nx.MultiGraph):
    multi_edges = [(u, v) for u, v, k in G.edges(keys=True) if k > 0]
    print(f"\nMultiple edges: {len(multi_edges)}")
else:
    print(f"\nGraph type: Simple graph (no multiple edges possible)")

# 4. Check connectedness
print(f"\nConnectivity:")
print(f"  Is connected: {nx.is_connected(G)}")
if not nx.is_connected(G):
    components = list(nx.connected_components(G))
    component_sizes = [len(c) for c in components]
    print(f"  Number of components: {len(components)}")
    print(f"  Largest component size: {max(component_sizes)} ({max(component_sizes)/G.number_of_nodes()*100:.1f}%)")
    print(f"  Smallest component size: {min(component_sizes)}")

# 5. Node degree distribution issues
zero_degree = [n for n in G.nodes() if G.degree(n) == 0]
print(f"\nDegree anomalies:")
print(f"  Nodes with degree 0: {len(zero_degree)}")
print(f"  Nodes with degree 1: {len([n for n in G.nodes() if G.degree(n) == 1])}")

# 6. Check for biases
print(f"\nPotential biases:")
print(f"  Sampling bias: Dataset is from limited ego-networks (not representative of all users)")
print(f"  Temporal bias: Snapshot in time (ignores network evolution)")
print(f"  Platform bias: Only captures Facebook connections (not other social media)")

print("="*70)
```
